```python
import os
import requests
import json
```

```python
import os
os.environ["GOOGLE_API_KEY"] = ""
os.environ["GEMINI_API_KEY"] = ""
```

```python

# --- 1. Configuration ---

# It is best practice to load the API key from an environment variable
# export GEMINI_API_KEY="YOUR_API_KEY" in your terminal
API_KEY = os.getenv("GEMINI_API_KEY")

if not API_KEY:
    raise ValueError("GEMINI_API_KEY environment variable not set.")

# The REST API endpoint for the generateContent method
API_ENDPOINT = "https://generativelanguage.googleapis.com/v1beta/models/gemini-2.5-flash:generateContent"
# Add the API key as a query parameter in the URL
URL = f"{API_ENDPOINT}?key={API_KEY}"

# --- 2. Request Body ---

# The JSON payload defining the prompt
prompt_text = "Write a short, fun poem about Python programming."

request_data = {
    "contents": [
        {
            "parts": [
                {
                    "text": prompt_text
                }
            ]
        }
    ]
}

# --- 3. Making the HTTP Request ---

# Set the required headers
headers = {
    "Content-Type": "application/json"
}

print(f"Sending request to: {URL}")

try:
    # Send the POST request
    response = requests.post(
        url=URL,
        headers=headers,
        data=json.dumps(request_data)
    )

    # Raise an exception for bad status codes (4xx or 5xx)
    response.raise_for_status()

    # Parse the JSON response
    response_json = response.json()

    # --- 4. Processing the Response ---

    # The generated text is typically found in the first candidate's part
    generated_text = response_json['candidates'][0]['content']['parts'][0]['text']

    print("\n--- Model Response ---")
    print(generated_text)
    print("----------------------")

except requests.exceptions.HTTPError as err:
    print(f"HTTP Error occurred: {err}")
    print(f"Response Body: {response.text}")
except Exception as e:
    print(f"An error occurred: {e}")
```
