```python
import requests
import json
import os
```

```python
import os
os.environ["GOOGLE_API_KEY"] = ""
os.environ["GEMINI_API_KEY"] = ""
```

```python


# Set your model and API key
MODEL_ID = "gemini-2.5-flash"
API_KEY = os.getenv("GEMINI_API_KEY") # Replace with your actual key or load from environment

# The URL for the Count Tokens endpoint (uses the public API)
COUNT_URL = f"https://generativelanguage.googleapis.com/v1beta/models/{MODEL_ID}:countTokens"

# Request body structure
request_data = {
    "contents": [
        {
            "role": "user",
            "parts": [
                {"text": "Explain the formula for kinetic energy to a beginner in one paragraph."}
            ]
        }
    ]
}

# Send the POST request
headers = {"Content-Type": "application/json"}
response = requests.post(
    COUNT_URL,
    headers=headers,
    params={"key": API_KEY},
    data=json.dumps(request_data)
)

# Parse and print the result
if response.status_code == 200:
    count_data = response.json()
    total_tokens = count_data.get("totalTokens", 0)
    print("--- Count Tokens API Result ---")
    print(f"Estimated Input Tokens: {total_tokens}")
else:
    print(f"Error: {response.status_code}\n{response.text}")
```
